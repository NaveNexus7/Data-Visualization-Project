```python
# Make a new data frame with only selected variables of interest from the dataset.
df_student_new_engagement = df_student_new9[['Sports','News', 'Fashion', 'Nightlife','Technology', 'Music']].copy() 

# Display the first five rows of the new data frame to inspect the selected variables.
df_student_new_engagement.head()
```

```python
# Define the columns of interest for engagement analysis
columns_of_interest = ['Sports', 'News', 'Fashion', 'Nightlife', 'Technology', 'Music']

# Create a list to store the data for analysis
data_to_append = []

# Iterate through the columns of interest
for column in columns_of_interest:
    # Iterate through the likelihood categories (-2, -1, 0, 1, 2)
    for likelihood in [-2, -1, 0, 1, 2]:
        # Filter the DataFrame based on the category and likelihood
        filtered_data = df_student_new_engagement[df_student_new_engagement[column] == likelihood]

        # Calculate the count of occurrences
        count = len(filtered_data)

        # Append the category, likelihood, and count to the data list
        data_to_append.append({'Category': column, 'Likelihood': likelihood, 'Count': count})

# Create a DataFrame from the list of dictionaries
counts_df_student_new_engagement = pd.DataFrame(data_to_append)

# Debugging: Print the resulting DataFrame for inspection
print(counts_df_student_new_engagement)
```

```python
# Pivot the DataFrame to reshape it based on engagement likelihood
pivoted_df = counts_df_student_new_engagement.pivot_table(index='Category', columns='Likelihood', values='Count', fill_value=0).reset_index()

# Rename the columns for better interpretation
pivoted_df.columns = ['Category', '-2', '-1', '0', '1', '2']
pivoted_df.reset_index(drop=True, inplace=True)

# Use the rename() function to update column names for better clarity in data visualization
pivoted_df.rename(columns={"-2": "Very unlikely", "-1": "Unlikely", "0": "Unsure", "1": "Likely", "2": "Very Likely"}, inplace=True)

# Display the reshaped DataFrame for inspection
print(pivoted_df)
```
